## Data Dictionary: tech_news_df

This DataFrame contains information scraped from articles in the HackerNoon HuggingFace dataset.

**Columns:**

* **Unnamed: 0:** *Index* - Original index from the source dataset.
* **companyName:** *string* - name of the Company HackerNoon dataset associated with the article.
* **companyUrl:** *string* - URL of the Company HackerNoon dataset associated with the article.
* **published_at:** *datetime* - Date and time the article was published.
* **url:** *string* - URL of the source article.
* **title:** *string* - Title of the source article.
* **text_path:** *string* - Path to text files containing scraped content of the article, or an error message. Note the filename "{i}.txt" where i corresponds to the value of the *Index* for that row

**Notes:**

* The dataset was obtained from Hugging Face's "HackerNoon/tech-company-news-data-dump" dataset.
* A subset of 80,000 articles was randomly sampled and used for analysis.
* The 'scraped_text' column was populated by scraping the full text from each article's URL.

```python
# read the text from the text_path for a random row of the DataFrame
# THIS IS RANDOM, SO EXECUTE THE CELL AGAIN TO SEE OTHER DATA/ARTICLE CONTENTS

# sample a data record of interest
row = tech_news_df.sample(1)
display(row)

# identify the source file
print("Scraped text in file: ", row.iloc[0].text_path)

# read the source file
with open(row.iloc[0].text_path, 'r') as f:
    text = f.read()
print("Scraped text: \n", text)
```
